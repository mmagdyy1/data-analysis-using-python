9: Given the 2D array mat:

a. Compute the mean of each row.

b. Compute the sum of each column.

c. Find the maximum value in the entire matrix and its position (row, column).

```python
mat = np.array([
    [4,  8,  1, 12],
    [7,  3,  9,  5],
    [11, 2,  6, 10],
    [15, 0, 14, 13]
])
mat
```

```python
#a
row_means = mat.mean(axis=1)
row_means
```

```python
#b
col_sums = mat.sum(axis=0)
col_sums
```

```python
#c
max_val = mat.max()
max_pos = np.unravel_index(mat.argmax(), mat.shape)
max_val
```
