import pandas library

```python
import pandas as pd
```

1. Load the file students.csv and make sure Score Date is parsed as datetime.

```python
df = pd.read_csv("students.csv", parse_dates=["Score Date"])
df.head()
```

2.Print the shape, columns list, and data types of the DataFrame.

```python
print("Shape:", df.shape)

print("Columns:")
print(df.columns)

print("Data Types:")
print(df.dtypes)
```

3.Show the first 7 rows and the last 3 rows

```python
#first 7 rows
print("First 7 rows:")
print(df.head(7))
#last 3 rows
print("Last 3 rows:")
print(df.tail(3))
```

4.Show how many missing values in each column

```python
print(df.isnull().sum())
```

5.What is the average, minimum, and maximum Grade?

```python
print("Average Grade:", df["Grade"].mean())
print("Minimum Grade:", df["Grade"].min())
print("Maximum Grade:", df["Grade"].max())
```

6.How many unique cities are there? List them

```python
print("Number of cities:", df["City"].nunique())

print("Cities:")
print(df["City"].unique())
```

7.Use value_counts() to show how many students per Subject

```python
print(df["Subject"].value_counts())
```

8.Select only the columns: Name, Grade, Subject, City

```python
df_selected = df[["Name", "Grade", "Subject", "City"]]

print(df_selected)
```

9.Show all students who Passed and have Grade ≥ 85

```python
result = df[(df["Passed"] == True) & (df["Grade"] >= 85)]

print(result)
```

10.Show students whose Name contains the letter "a" or "A" (case insensitive)

```python
result = df[df["Name"].str.contains("a", case=False)]

print(result)
```

11.Show students in Cairo or Alexandria with Grade < 80

```python
result = df[(df["City"].isin(["Cairo", "Alexandria"])) & (df["Grade"] < 80)]

print(result)
```

12.Show students with Grade between 80 and 90 (inclusive)

```python
result = df[df["Grade"].between(80, 90)]

print(result)
```

13.Using .loc, change Khaled's grade from 55 to 68

```python
df.loc[df["Name"] == "Khaled", "Grade"] = 68
print(df[df["Name"] == "Khaled"])
```

14.Add a new column Is_Excellent that is True if Grade ≥ 90, False otherwise

```python
df["Is_Excellent"] = df["Grade"] >= 90
print(df[["Name", "Grade", "Is_Excellent"]])
```

15.Fill missing Grade values with the average Grade

```python
avg_grade = df["Grade"].mean()

df["Grade"].fillna(avg_grade, inplace=True)
print(df)
```

16.Fill missing Score Date values with the most common date (mode)

```python
mode_date = df["Score Date"].mode()[0]

df["Score Date"].fillna(mode_date, inplace=True)
print(df)
```

17.Add a new column Final_Grade = Grade + 5

```python
df["Final_Grade"] = df["Grade"] + 5
print(df[["Name", "Grade", "Final_Grade"]])
```

18.Create a column Grade_Category using pd.cut with bins: 
•0–69 → "Fail" 
•70–79 → "Average" 
•80–89 → "Good" 
•90–100 → "Excellent" 

```python
bins = [0, 69, 79, 89, 100]

labels = ["Fail", "Average", "Good", "Excellent"]

df["Grade_Category"] = pd.cut(df["Grade"], bins=bins, labels=labels)
print(df[["Name", "Grade", "Grade_Category"]])
```

19.Remove the column Student ID

```python
df.drop("Student ID", axis=1, inplace=True)
print(df)
```

20.Keep only students who Passed

```python
df = df[df["Passed"] == True]
print(df)
```

21.Sort the DataFrame by Grade descending

```python
df_sorted = df.sort_values(by="Grade", ascending=False)

print(df_sorted)
```

22.Sort by City ascending, then inside each city sort by Grade descending

```python
df_sorted = df.sort_values(by=["City", "Grade"], ascending=[True, False])

print(df_sorted)
```

23.Add a column Rank that ranks students by Grade

```python
df["Rank"] = df["Grade"].rank(method="min", ascending=False)
print(df[["Name", "Grade", "Rank"]])
```

24.Show the top 4 students with the highest grades

```python
top_students = df.nlargest(4, "Grade")

print(top_students)
```

25.Show the 3 youngest students

```python
youngest = df.nsmallest(3, "Age")

print(youngest)
```

26.Find the student(s) with the highest Grade and show their Name and Grade

```python
max_grade = df["Grade"].max()

top_students = df[df["Grade"] == max_grade][["Name", "Grade"]]

print(top_students)
```

27.After filling missing grades, recalculate the average Grade per Subject

```python
avg_subject = df.groupby("Subject")["Grade"].mean()

print(avg_subject)
```

28.Create a new column City_Avg using transform

```python
df["City_Avg"] = df.groupby("City")["Grade"].transform("mean")
print(df[["Name", "City", "Grade", "City_Avg"]])
```

29.Show the subject with the highest average grade

```python
avg_subject = df.groupby("Subject")["Grade"].mean()

highest_subject = avg_subject.idxmax()

print("Subject with highest average:", highest_subject)
```
